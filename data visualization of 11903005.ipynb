```python
import pandas as pd,numpy as np,matplotlib as mt,matplotlib.pyplot as pit,matplotlib.style as style
from matplotlib import rc
from sklearn.preprocessing import LabelEncoder
style.use('ggplot')
df=pd.read_excel(r'D:\DS\11903005.xlsx')
```

# Camparing BMI of students lives in hostels and Day Scholar(scatterplot )


```python
df['BMI']=df['Weight']/((df['Height']/100)**2)
#df.drop_duplicates(subset ="Regd No", keep = False, inplace = True)
new_df=df[['Regd No','Gender','ScholarType','BMI']].drop_duplicates()#new_df is contain all the values 
ghseries=new_df[(new_df.ScholarType == "Hostler") & (new_df.Gender == "Female")]
gsseries=new_df[(new_df.ScholarType == "Day Scholar") & (new_df.Gender == "Female")]
x= ghseries['Regd No']
y= ghseries['BMI']
x2=gsseries['Regd No']
y2=gsseries['BMI']
rc('font',weight='bold')
pit.figure(figsize=(13,8))
pit.scatter(x,y,label='Hostler',color='#239B56',edgecolors='g',s=150)
pit.scatter(x2,y2,label='Day Scholar',color='#B7950B',edgecolors='g',s=150)
pit.xlabel('BMI')
pit.ylabel('Regd No')
pit.title('Female BMI comparison')
#pit.grid(True,color="#BA4A00")
pit.legend(bbox_to_anchor=(1.2, 1.05))
pit.show()
```

```python
Mhseries=new_df[(new_df.ScholarType == "Hostler") & (new_df.Gender == "Male")]
Msseries=new_df[(new_df.ScholarType == "Day Scholar") & (new_df.Gender == "Male")]
x3= Mhseries['Regd No']
y3= Mhseries['BMI']
x4=Msseries['Regd No']
y4=Msseries['BMI']
rc('font',weight='bold')
pit.figure(figsize=(13,8))
pit.scatter(x3,y3,label='Hostler',color='#1ABC9C',edgecolors='k',s=150)
pit.scatter(x4,y4,label='Day Scholar',color='#3498DB',edgecolors='r',s=150)
pit.xlabel('BMI')
pit.ylabel('Regd No')
pit.title('Male BMI comparison')
#pit.grid(True,color="#7B241C")
pit.legend(bbox_to_anchor=(1.2, 1.05))
pit.show()
```

# Study Medium Distribution over Direction Distribution(stacked barplot)
Ratio_proporation_of_Medium(pie chart)

```python
pie=df[['Regd No','Medium']].drop_duplicates().Medium
r=0
e=0
h=0
for x in pie:
    if x == "Regional":
        r=r+1
    elif x== "Hindi":
        h=h+1
    elif x== "English":
        e=e+1
    else :
        pass
Medium_list=[r,e,h]
activities = ['Regional','English','Hindi']
c=['#f95d6a','#ff7c43','#ffa600']
explodeTuple = (0.040,0.040,0.040)
rc('font',weight='bold')
pit.pie(Medium_list, labels=activities, startangle=120, autopct='%.2f%%',explode=explodeTuple,shadow=True,radius=2.0,colors=c)
pit.legend(bbox_to_anchor=(1.5, 1.55))
pit.show()    
```

```python
bar=df[['Regd No','Direction','Medium']].drop_duplicates()
West=bar[bar.Direction == "West"].Medium
North=bar[bar.Direction == "North"].Medium
South=bar[bar.Direction == "South"].Medium
East=bar[bar.Direction == "East"].Medium
#ths function will calculate the number of students in each medium according to direction
def mid(x):
    R=0
    H=0
    E=0
    for a in x:
        if a == "Regional":
            R=R+1
        elif a == "Hindi":
            H=H+1
        elif a == "English":
            E=E+1
        else :
            pass
    y=[R,H,E]
    return y
q=mid(West)
e=mid(East)
s=mid(South)
n=mid(North)
#controle the style
rc('font', weight='bold')
 # Values of each group
bars1 = n
bars2 = s
bars3 = q
bars4 = e
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
barrs = np.add(bars, bars3).tolist()
r = [0,1,2]
names = ['Regional','English','Hindi']
barWidth = 0.7
pit.figure(figsize=(13,8))
pit.bar(r, bars1, color='#004c6d', edgecolor='#004c6d', width=barWidth)
pit.bar(r, bars2, bottom=bars1, color='#346888', edgecolor='#004c6d', width=barWidth)
pit.bar(r, bars3, bottom=bars, color='#7aa6c2', edgecolor='#004c6d', width=barWidth)
pit.bar(r, bars4, bottom=barrs, color='#c1e7ff', edgecolor='#004c6d', width=barWidth)
pit.xticks(r, names, fontweight='bold')
pit.xlabel("Medium")
pit.ylabel('Number of Students')
pit.title('Study Medium Distribution')
pit.legend(('North', 'South','West', 'East'))
pit.show() 
```

# Effect of Attendance on Grades
